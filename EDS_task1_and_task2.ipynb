```python
import pandas as pd
import operator
import plotly.express as px
%matplotlib inline
```

# Data set

```python
path='https://covid.ourworldindata.org/data/owid-covid-data.csv'
```

```python
all_countries=pd.read_csv(path,sep=',')
```

```python
#all_countries.columns   ##to know all the columns in the data set
```

```python
all_countries.to_csv('covid_new_data_owid.csv',';')
```

```python
data_Ger=all_countries[all_countries['location']=='Germany']
```

```python
data_Ger=data_Ger.reset_index(drop=True)
```

```python
#data_Ger
```

```python
## data_Ger.columns
```

# Plotting for a single country

```python
## data_Ger.head(5)
```

```python
data_Ger['total_cases_relative']=data_Ger['total_cases']/data_Ger['population']
```

```python
fig = px.line(data_Ger, x="date", y="total_Number_of_cases", title='Covid cases in Germany')
fig.show()
```

```python
fig = px.line(data_Ger, x="date", y= "Relative_number_of_covid_cases", title='relative Covid cases in Germany ')
fig.show()
```

# plotting with 3 countries

```python
## all_countries['location'].unique()   ##to know all the countries
```

```python
d_Ger=all_countries['location']=='Germany'
d_France=all_countries['location']=='France'
d_Brazil=all_countries['location']=='Brazil'
```

```python
d_France =all_countries['location']=='France'
```

```python
d_Brazil=all_countries['location']=='Brazil'
```

```python
#sum(d_US)
```

```python
##sum(d_Ger|d_Ind|d_US)
```

```python
d_all=d_Ger|d_France|d_Brazil
```

```python
data_all_countries=all_countries[d_all]
```

```python
data_all_countries=data_all_countries.reset_index(drop=True)
```

```python
#data_all_countries.columns
```

```python
fig = px.line(data_all_countries, x="date", y="total_cases", color='location', title='Covid cases in 3 countries')
fig.show()
```

```python
data_all_countries['total_cases_relative']=data_all_countries['total_cases']/data_all_countries['population']
```

```python
fig = px.line(data_all_countries, x="date", y="total_cases_relative", color='location', title='Covid cases in all 3 countries')
fig.show()
```

# Vaccinations

```python
fig = px.line(data_all_countries, x="date", y="total_vaccinations", color='location', title='Total vaccinations in all 3 countries')
fig.show()
```

# Vaccinations with Interpolated data

```python
data_all_countries['total_vaccinations_intp'] = data_all_countries['total_vaccinations'].interpolate(method="pad", limit=2)

#data_all_countries.columns
```

```python
fig = px.line(data_all_countries, x="date", y='total_vaccinations_intp', color='location', title='Total vaccinations in all 3 countries')
fig.show()
```
